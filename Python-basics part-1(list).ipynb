```python
# Create a new list that contains only the even numbers from the original list.

numbers =[]
mylist =[2,4,5,7,56,23,59,73,0]

for i in mylist:
    if i%2==0:
        numbers.append(i)
    else:
        pass

print("this is your even numbers:", numbers)        
```

```python
# Reverses the list(by pop function).

mylist =[2,4,5,7,56,23,59,73,0]
for i in range(len(mylist)):
    mylist.insert(i,mylist.pop())
print(mylist)    
                   
```

```python
# Reverse a list by slicing operator.

mylist =[2,4,5,7,56,23,59,73,0]
print(mylist[::-1])
```

```python
# Sort the list without using the sort operator(slicing operator).

mylist =[2,4,5,7,56,23,59,73,0]
for i in range(len(mylist)):
    for j in range(i+1,len(mylist)):
        if mylist[i]>=mylist[j]:
            mylist[i],mylist[j] = mylist[j],mylist[i]
print(mylist)            
```
